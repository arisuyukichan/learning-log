**C6.** Keep prompting the user for a number until they enter 'done'. Print each number doubled. (Pseudo-code in notebook; for real, do this in a .py file.)

```python
while True:
    user_input = input("Enter a number (or 'done' to quit): ")

    if user_input.strip().lower() == 'done':
        print("Goodbye!")
        break

    try:
        number = float(user_input)

        doubled = number * 2

        if doubled.is_integer():
            print(f"Doubled: {int(doubled)}")
        else:
            print(f"Doubled: {doubled}")

    except ValueError:
        print("Invalid input. Please enter a valid number or 'done'.")
```

Given a list of word entries (dicts with `word`, `ipa`, `thai`), return a new list containing only entries where the word is ≥ 8 letters.

```python
WORDS: list[dict[str, str]] = [
    {"word": "thorough",   "ipa": "/ˈθʌrə/",        "thai": "ละเอียด"},
    {"word": "ubiquitous", "ipa": "/juːˈbɪkwɪtəs/", "thai": "พบเห็นได้ทั่วไป"},
    {"word": "cat",        "ipa": "/kæt/",          "thai": "แมว"},
]
# Your code: build `long_words` — list of entries with word length >= 8
long_words = [entry for entry in WORDS if len(entry["word"]) >= 8]

print(long_words)
```

**E2. Vowel counter.**

Write a `for` loop that counts how many vowels (a/e/i/o/u) are in the word `"ubiquitous"`. Print the count.

```python
word = "ubiquitous"
# Your code, expected output: Vowels: 5
vowels = "aeiou"
vowel_count = 0

for letter in word:
    if letter in vowels:
        vowel_count += 1

print(vowel_count)
```

**E4. Word counter dict.**

Given a list of words (some repeated), build a dict mapping `word → count`. E.g., `['the', 'cat', 'the', 'dog']` → `{'the': 2, 'cat': 1, 'dog': 1}`. (You can use `dict.get(key, default)` for the common pattern.)

```python
words = ['the', 'cat', 'sat', 'on', 'the', 'mat']
# Your code: build `counts`
word_counts = {}

for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

print(word_counts)
```

**E5. Multi-round quiz logic.**

Write a `while` loop that picks a random word from `WORDS`, asks the user 5 times total (no more, no less). Track `correct` count. Print final score. (Mock the input by hardcoding answers in your draft — don't run input() in a notebook.)

```python
import random
WORDS = [
    {'word': 'cat', 'ipa': '/kæt/'},
    {'word': 'dog', 'ipa': '/dɒg/'},
    {'word': 'fish', 'ipa': '/fɪʃ/'},
]
# Sketch the 5-round quiz logic. Use a `for` loop OR a `while` with a counter.
mock_answers = ['cat', 'dog', 'bird', 'fish', 'cow']
mock_index = 0

correct_count = 0
attempts = 0

while attempts < 5:
    chosen_entry = random.choice(WORDS)
    correct_word = chosen_entry['word']
    ipa = chosen_entry['ipa']

    print(f"Round {attempts + 1}: What word is pronounced as {ipa}?")

    user_answer = mock_answers[mock_index]
    mock_index += 1
    print(f"Your answer (mocked): {user_answer}")

    if user_answer.strip().lower() == correct_word:
        print("Correct")
        correct_count += 1
    else:
        print(f"Incorrect. The correct answer was '{correct_word}'.")
```

**E6. Skip the easy ones.**

Given the WORDS list, build a quiz that ONLY asks about words >= 7 letters long. Use `continue` to skip short words inside the loop.

```python
WORDS = [
    {'word': 'cat', 'ipa': '/kæt/'},
    {'word': 'thorough', 'ipa': '/ˈθʌrə/'},
    {'word': 'a', 'ipa': '/ə/'},
    {'word': 'ubiquitous', 'ipa': '/juːˈbɪkwɪtəs/'},
]
# Your code: for each entry, skip if len(word) < 7, else 'print' (mock the quiz)
mock_answers = ['thorough', 'ubiquitous']
mock_index = 0

correct_count = 0
total_asked = 0

for entry in WORDS:
    if len(entry['word']) < 7:
        continue

    correct_word = entry['word']
    ipa = entry['ipa']

    print(f"What word is pronounced as {ipa}?")

    user_answer = mock_answers[mock_index]
    mock_index += 1
    print(f"Your answer (mocked): {user_answer}")

    if user_answer.strip().lower() == correct_word:
        print("Correct")
        correct_count += 1
    else:
        print(f"Incorrect. The correct answer was '{correct_word}'.")

    print("-" * 30)
    total_asked += 1

print(f"Quiz finished! You scored {correct_count} out of {total_asked}.")
```

**E7. Word frequency from a sentence.**

Given a sentence, return a dict of word → count. Lowercase everything first. Use `.split()` to break into words.

```python
sentence = 'The quick brown fox jumps over the lazy dog the fox'
# Your code
lowered_sentence = sentence.lower()
words = lowered_sentence.split()
word_counts = {}
for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

print(word_counts)
```
