**E6. English Helper warm-up — sentence with word count.**

Given `sentence = "Smartphones are now ubiquitous in modern life"`, print:
```
Sentence: Smartphones are now ubiquitous in modern life
Word count: 7
```
(Hint: a sentence's word count = number of spaces + 1. Or use `sentence.count(" ") + 1`.)

````python
sentence = "Smartphones are now ubiquitous in modern life"
# Your code
word_count = len(sentence.split())
print(f"Sentence: {sentence}\nWord count: {word_count}")
````

**E7. Final challenge.**

A new English word costs you 30 seconds to learn. You want to learn 500 words. How many **hours** is that? Print to 2 decimal places.

````python
words_to_learn = 500
seconds_per_word = 30
# Your code — print hours, 2 decimals
total_seconds = words_to_learn * seconds_per_word
total_hours = total_seconds / 3600

print(f"{total_hours:.2f}")
````
