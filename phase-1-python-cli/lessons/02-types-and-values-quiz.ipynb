**E5.** Write a function `card_summary(entry)` that takes a word dict and returns a one-line summary like `"resilient (/rɪˈzɪliənt/) — adjective — ยืดหยุ่น"`. Type-hint the parameter and return.

```python
# Your code
def card_summary(entry: dict[str, str]) -> str:
    word = entry.get("word", "N/A")
    ipa = entry.get("ipa", "N/A")
    pos = entry.get("pos", "N/A")
    thai = entry.get("thai", "N/A")
    return f"{word} ({ipa}) — {pos} — {thai}"

# Demonstration
example_card = {"word": "resilient", "ipa": "/rɪˈzɪl.jənt/", "thai": "ยืดหยุ่น", "pos": "adjective"}
print(card_summary(example_card))

# Using a card from the word_card list (E2)
if 'word_card' in locals() and len(word_card) > 0:
    print(card_summary(word_card[0]))
```

**E6.** Write a typed function `count_long_words(words, min_length=8)` that returns how many words in the list are at least `min_length` letters long.

```python
# Your code
def count_long_words(words: list[dict[str, str | int]], min_length: int = 8) -> int:
    count: int = 0
    for word_entry in words:
        word: str = str(word_entry.get("word", "")) # Ensure 'word' is treated as string for len()
        if len(word) >= min_length:
            count += 1
    return count

# Demonstration
print(f"Number of long words (>=8 chars): {count_long_words(word_card)}")
print(f"Number of words (>=10 chars): {count_long_words(word_card, min_length=10)}")
```

**E7.** Write a typed function `find_word(words, target)` that returns the word entry matching `target`, or `None` if not found. Use `dict[str, str] | None` as the return type.

```python
# Your code
def find_word(words: list[dict[str, str | int]], target: str) -> dict[str, str | int] | None:
    for entry in words:
        if entry.get("word") == target:
            return entry
    return None

# Demonstration
print(f"Find 'Anachronism': {find_word(word_card, 'Anachronism')}")
print(f"Find 'NonExistentWord': {find_word(word_card, 'NonExistentWord')}")
```
